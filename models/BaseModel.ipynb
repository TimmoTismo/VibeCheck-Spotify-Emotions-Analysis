```python
# Importing necessary libraries
from sklearn.model_selection import train_test_split
import pandas as pd
import math
```

```python
# Read in the playlists dataset
data = pd.read_csv('../datasets/data.csv')
data
```

```python
# Pre-process the data

# Remove the index column from csv file
data = data.loc[:, ~data.columns.str.contains('^Unnamed')] 

# Drop unnecessary columns
df = data.drop(columns=['duration_ms', 'nlp_lyrics', 'nlp_annotations', 'time_signature', 'valence+nlp'	])

# Randomise rows
df = df.sample(frac=1).reset_index(drop=True)

df
```

```python
X = df.iloc[:, 2:13] # All rows, features only, no labels
y = df.iloc[:, 13] # All rows, label only, no features
```

```python
# Dividing X, y into train and test data
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
```
